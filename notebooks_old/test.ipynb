```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import qmc
```

```python
lh = qmc.LatinHypercube(2, seed=23548709)
smpl = lh.random(400)
smpl = qmc.scale(smpl, (-0.5,-0.5), (1,1))
X1 = smpl[:,0]
Y1 = smpl[:,1]

np.stack((X1, Y1), axis=1).shape
```

```python
# X1, Y1 = np.meshgrid(tmp1, tmp2)    
Z1 = (0.75*np.exp(- (9*X1 - 2)**2/4 - (9*Y1 - 2)**2/4) + 
        0.75*np.exp(- (9*X1 + 1)**2/49 - (9*Y1 + 1)/10) +
        0.5*np.exp(- (9*X1 - 7)**2/4  - (9*Y1 - 3)**2/4) +
        0.2*np.exp(- (9*X1 - 4)**2    - (9*Y1 - 7)**2))
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(6,6), subplot_kw={'projection': '3d'})
ax.patch.set_alpha(0)
ax.plot_trisurf(X1, Y1, Z1, cmap='viridis')
```
